```python
import numpy as np
import pandas as pd
```

```python
dicts = {
    'text' : ['people watch campusx' , 'campusx watch campusx' , 'people write comment' , 'campusx write comment'],
    'output' :[1,1,0,0]
}
```

```python
df = pd.DataFrame(dicts)
```

```python
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
```

```python
bow = cv.fit_transform(df['text'])
```

```python
cv.vocabulary_
```

```python
bow[0].toarray()
```

```python
bow[1].toarray()
```

```python
cv.transform(['campusx of watch hello campusx the comment ']).toarray()
# oov is handled they are not counted 
# we have to pass a list or a dataframe as document is needed to be passed
```

 #### using n grams using hyperparameter in count vectoriser


```python
cv = CountVectorizer(ngram_range = (1,3))
```

```python
n_grams = cv.fit_transform(df['text'])
```

```python
cv.vocabulary_
```

```python
n_grams[0].toarray()
```

```python
n_grams[1].toarray()
```
