```python
import re
```

```python
sample_text = """
<p>
This is the <strong>first line</strong> with <em>emphasis</em> and a
<a href="#">link</a>.<br>
This is the <span style="color:blue;">second line</span> with
<mark>highlighted</mark> text.
</p>
"""
```

```python
def striptml(data):
    p = re.compile(r'<.*?>')
    return p.sub('' , data)
```

```python
striptml(sample_text)
```
